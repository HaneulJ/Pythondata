```python
from matplotlib import font_manager, rc
font_path = "data/THEdog.ttf"   #폰트파일의 위치
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

import pandas as pd
import matplotlib.pyplot as plt 
import matplotlib
import numpy as np
```

```python
#1. 데이터프레임을 생성하여 df변수에 저장하고 출력

df = pd.DataFrame([[500, 450, 520, 610], [690, 700, 820, 900],
                    [1100, 1030, 1200, 1380], [1500, 1650, 1700, 1850],
                    [1990, 2020, 2300, 2420], [1020, 1600, 2200, 2550]], 
                   index=[2015, 2016, 2017, 2018, 2019, 2020],
                   columns=['1분기', '2분기', '3분기', '4분기'])
display(df)
```

```python
#2.    
df.index = df.index.map(lambda x : str(x) + "년")
plt.figure(figsize=(10,6))
plt.plot(df)
plt.title('시각화 과제1', color="purple", fontsize=20)
plt.legend(df)

plt.savefig("output/hw1.png") 
plt.show()
```

```python
#3. 
plt.figure(figsize=(10,6))
plt.plot(df.T)
plt.title('시각화 과제2', color="red", fontsize=20)
plt.legend(df.T)

plt.savefig("output/hw2.png") 
plt.show()
```

```python
#4.
plt.figure(figsize=(10,6))
ratio = [20.8, 22.8, 26.7, 29.7]
labels = ['1분가', '2분기', '3분기', '4분기']
explode = [0.05, 0.05, 0.05, 0.05]

plt.pie(ratio, labels=labels, autopct='%.1f%%', explode=explode, shadow=True)
plt.savefig("output/hw3.png") 

plt.show()
```

```python
#5.
plt.figure(figsize=(10,6))
ratio = [6.4, 9.5, 14.4, 20.5, 26.7, 22.5]
labels = ["2015년", "2016년", "2017년", "2018년", "2019년", "2020년"]

plt.pie(ratio, labels=labels, autopct='%.1f%%')
plt.savefig("output/hw4.png") 
plt.show()
```
