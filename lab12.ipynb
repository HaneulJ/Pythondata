```python
#15. df 에서 직원이름, 월급 그리고 커미션 정보만으로 구성된 데이터프레임을 생성하고 df1 변수에 저장하고 앞에부터 5개 행만 출력
df1 = df[["ename","sal","comm"]]
display(df1.head(5))
```

```python
#17. df1 을 가지고 다음과 같이 시각화 결과를 출력

df1 = df1.T
df1.plot(kind='bar', color=["green", "yellow"])
```
