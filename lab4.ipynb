```python
# https://comic.naver.com/genre/bestChallenge.nhn 페이지의 웹툰정보를 추출해서 title, summary  그리고 rating 정보를 각각 출력한다.
# 10개 페이지까지 진행한다

import requests
from bs4 import BeautifulSoup
import re

for n in range(1,11): 
    req = requests.get('https://comic.naver.com/genre/bestChallenge.nhn?&page=' + str(n))
    html = req.text 
    bs = BeautifulSoup(html, 'html.parser')

    titles = bs.select("h6 > a")

    title = []

    for i in titles:
        content = i.text
        content = re.sub("[\n\t]","", content)
        content = re.sub("  ","", content)
        title.append(content) 
    
    print(title)

for n in range(1,11): 
    req = requests.get('https://comic.naver.com/genre/bestChallenge.nhn?&page=' + str(n))
    html = req.text 
    bs = BeautifulSoup(html, 'html.parser')

    summaries = bs.select("div.summary")

    summary = []

    for i in summaries:
        summary.append(i.text)

    print(summary)
    
    
for n in range(1,11): 
    req = requests.get('https://comic.naver.com/genre/bestChallenge.nhn?&page=' + str(n))
    html = req.text 
    bs = BeautifulSoup(html, 'html.parser')

    ratings = bs.select("div.rating_type > strong")

    rating = [] 

    for i in ratings:
        rating.append(i.text)
    
    print(rating)
```
